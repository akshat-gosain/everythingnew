```python
import matplotlib.pyplot as plt
import numpy as np

def midpoint(p1, p2):
    """
    Calculates the midpoint between two points.
    Args:
        p1 (np.array): A point in space.
        p2 (np.array): Another point in space.
    Returns:
        np.array: The midpoint between p1 and p2.
    """
    return (p1 + p2) / 2

def chaos_game(vertices, iterations, transient=500):
    """
    Generates points for the Sierpinski Triangle using the Chaos Game method.
    Args:
        vertices (list of np.array): List of vertices (3 points) of the triangle.
        iterations (int): Number of iterations.
        transient (int): Number of transient values to skip (default: 500).
    Returns:
        np.array: Array of points representing the Sierpinski Triangle.
    """
    num_vertices = len(vertices)
    points = [np.random.rand(2)]  # Initial random point
    for _ in range(iterations):
        current_point = points[-1]
        random_vertex = vertices[np.random.randint(num_vertices)]
        new_point = midpoint(current_point, random_vertex)
        points.append(new_point)
    return np.array(points[transient:])

def plot_sierpinski(vertices, iterations, transient=500):
    """
    Plots the Sierpinski Triangle.
    Args:
        vertices (list of np.array): List of vertices (3 points) of the triangle.
        iterations (int): Number of iterations.
        transient (int): Number of transient values to skip (default: 500).
    """
    sierpinski_points = chaos_game(vertices, iterations, transient)
    plt.scatter(sierpinski_points[:, 0], sierpinski_points[:, 1], s=1, c='black')
    plt.title("Sierpinski Triangle")
    plt.axis('off')  # Hide axes
    plt.show()

if __name__ == '__main__':
    # Define vertices of the equilateral triangle
    vertex_a = np.array([0, 0])
    vertex_b = np.array([1, 0])
    vertex_c = np.array([0.5, np.sqrt(3) / 2])
    triangle_vertices = [vertex_a, vertex_b, vertex_c]

    # Set the number of iterations (adjust as needed)
    total_iterations = 10000

    # Plot the Sierpinski Triangle
    plot_sierpinski(triangle_vertices, total_iterations)
```
